```python
# 최고 기온을 실수로 변환 2
f = open('seoul.csv') 
data = csv.reader(f)
header = next(data)
print(header)
for row in data:
    if row[-1] =='':
        row[-1] = -999 # -999를 넣어 빈 문자열이 있던 자리라고 표시  
    row[-1] = float(row[-1]) 
    print(row) 
f.close()
```

### *서울의 기온이 가장 높았던 날의 날짜와 기온 구하기 

```python
import csv 
f = open('seoul.csv') 
data = csv.reader(f)
header = next(data)
max_temp = -999 
max_date = '' 
for row in data:
    if row[-1] =='':
        row[-1] = -999 # -999를 넣어 빈 문자열이 있던 자리라고 표시  
    row[-1] = float(row[-1])
    if max_temp < row[-1]:
        max_date = row[6]
        max_temp = row[-1] 
f.close()
print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은', max_date+ '로', max_temp, '도 였습니다.')
```

```python
import matplotlib.pyplot as plt 
plt.title('plotting') #여러개 연속되는 점 
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
#그래프에 범례 넣기 (설명)
plt.title('legend')
plt.plot([10, 20, 30, 40], color = 'skyblue', label = 'asc')
plt.plot([40, 30, 20, 10], 'pink', label = 'desc')
plt.legend()
plt.show()
```

```python
#선 모양 바꾸기 
plt.title('linestyle') 
plt.plot([10, 20, 30, 40], color = 'r' , linestyle = '--', label = 'dashed')
plt.plot([40, 30, 20, 10], color = 'g' , ls = ':' , label = 'dotted')
plt.legend()
plt.show()
```

```python
#마커 모양 바꾸기 
plt.title('marker') 
plt.plot([10, 20, 30, 40], 'r' , linestyle = '--', label = 'dashed')
plt.plot([40, 30, 20, 10], 'g^' , label = 'triangle up')
plt.legend()
plt.show()
```
