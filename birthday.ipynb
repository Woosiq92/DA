```python
# 데이터 리스트에 저장하기 
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f) 
next(data)
result = [] # 최고 기온 데이터를 저장할 리스트 생성 

for row in data:
    if row[-4] != '': 
        result.append(float(row[-4]))
print(result) 
print(len(result))
plt.plot(result, 'r')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f) 
next(data) 
high = [] # 최고 기온 값을 저장할 리스트 high 생성 
low = [] # 최저 기온 값을 저장할 리스트 low 생성 

for row in data : 
    if row[-4] != '' and row[-1] != '': 
        if 1983 <= int(row[-3].split('-')[0]):
            if row[-3].split('-')[1] == '02':
                high.append(float(row[-4]))
                low.append(float(row[-1]))
                
plt.rc('font', family = 'Malgun Gothic')
plt.title('내 생일의 기온 변화 그래프') 
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지 
plt.plot(high, 'hotpink', label = 'high')
plt.plot(low, 'skyblue', label = 'low')
plt.legend()
plt.show()
f.close()
```
